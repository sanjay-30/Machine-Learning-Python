```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#loading the data
food=pd.read_csv('food.tsv', sep='\t')
food.head()
```

```python
# storing the original data in a variable
food_original=food.copy()
```

```python
food.shape
```

```python
food.describe()
```

```python
#structure of data
food.info()
```

```python
#checking for null values
nulldata=food.isnull().sum().sort_values(ascending=False)
```

```python
#plotting the null values from highest to lowest
plt.figure(dpi=800)
nulldata.plot(kind='bar', title='Percentage of missing values per feature', figsize=(15,5),
             color='red', fontsize=5)
```

```python
#now we will drop the columns that contains only NaN

nulldata.index
```

```python
nulldata.head()
```

```python
len(nulldata)
```

```python
nulldata=nulldata/food.shape[0]*100
```

```python
nulldata.head()
```

```python
useless_features=nulldata[nulldata==100].index
print(useless_features)
```

```python
len(useless_features) #so there are 16 columns with only NaN data
```

```python
#Drop the useless features

food.drop(useless_features, axis=1, inplace=True)
print( food.shape)      #the columns are reduced from 163->147
```

```python
#still there ar NaN values so let us try to remove that also

high_nan_features=nulldata[(nulldata>50) & (nulldata<=100)]
print('High NaN features:', high_nan_features)
```

```python
len(high_nan_features)
```

```python
#we didn't use drop fun directly cause of attribute mismath
for i in high_nan_features.index:
    if i in food.columns:
        food.drop(i, axis=1, inplace=True)
# Now printing the shape of the food dataset
print( food.shape)
```

```python
# dropping nan values from the food dataset
food.dropna(inplace=True)
print('Shape of the food dataset after dropping NaN values is:', food.shape)
```

```python
#New data after cleaning
food.info()
```

```python

# Most occurs products are
food.product_name.value_counts().head(10)
```

```python
# Least occurs products are
food.product_name.value_counts().tail(10)
```

```python
#dropping duplicates
food.drop_duplicates(inplace=True)
print( food.shape)
```

```python
plt.figure(figsize=(10,5), dpi=150)
sns.heatmap(food.corr(), annot=True)
```
